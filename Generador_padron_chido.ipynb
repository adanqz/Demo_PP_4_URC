```python
import hashlib
import random
import time
from datetime import datetime, timedelta
from faker import Faker

fake = Faker('es_MX')

def generate_dataset():
    dataset = []
    current_date = datetime.now()
    id_counter = 0
    previous_hash = ""

    def generate_full_name(gender):
        if gender == "M":
            first_name = fake.first_name_male()
            if random.random() < 0.5:
                second_name = ""
            else:
                second_name = fake.last_name_male()
        else:
            first_name = fake.first_name_female()
            if random.random() < 0.5:
                second_name = ""
            else:
                second_name = fake.last_name_female()

        last_name1 = fake.last_name()
        last_name2 = fake.last_name()
        full_name = f"{first_name} {second_name} {last_name1} {last_name2}"

        return full_name

    for _ in range(10000):
        try:
            gender = random.choice(["M", "F"])  # Generar género aleatoriamente

            full_name = generate_full_name(gender)
            id_hash = hashlib.md5(full_name.encode()).hexdigest()
            phone_number = "5255" + str(random.randint(0, 999999)).zfill(6)
            public_hash = hashlib.sha256(full_name.encode()).hexdigest()
            private_hash = hashlib.sha512(full_name.encode()).hexdigest()
            value = random.randint(1, 5)
            timestamp = random.randint(0, int(time.mktime(current_date.timetuple())))
            timestamp = datetime.fromtimestamp(timestamp).strftime('%Y%m%d')
            clave_elector = str(random.randint(0, 9999999999999)).zfill(13)

            # Generar código encriptado que depende del registro anterior
            data_to_hash = f"{previous_hash}{id_hash}{phone_number}{public_hash}{private_hash}{value}{timestamp}{clave_elector}"
            encrypted_hash = hashlib.sha256(data_to_hash.encode()).hexdigest()

            data = {
                "id": id_counter,
                "name": full_name,
                "id_hash": id_hash,
                "phone": phone_number,
                "public_hash": public_hash,
                "private_hash": private_hash,
                "value": value,
                "timestamp": timestamp,
                "clave_elector": clave_elector,
                "encrypted_hash": encrypted_hash
            }
            dataset.append(data)

            previous_hash = encrypted_hash  # Actualizar el valor del hash anterior
            id_counter += 1
        except Exception as e:
            print(f"Error al generar los datos: {str(e)}")

    return dataset

dataset = generate_dataset()

for data in dataset[:99]:
    print(data)
```
