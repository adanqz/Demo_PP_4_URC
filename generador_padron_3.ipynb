```python
import hashlib
import random
import time
from datetime import datetime, timedelta
from faker import Faker

fake = Faker('es_MX')

def generate_dataset():
    dataset = []
    current_date = datetime.now()
    id_counter = 0

    def generate_full_name(gender):
        if gender == "M":
            first_name = fake.first_name_male()
            if random.random() < 0.5:
                second_name = ""
            else:
                second_name = fake.first_name_male()
        else:
            first_name = fake.first_name_female()
            if random.random() < 0.5:
                second_name = ""
            else:
                second_name = fake.first_name()

        return f"{first_name} {second_name}"

    for _ in range(10000):
        gender = random.choice(["M", "F"])  # Generar género aleatoriamente

        full_name = generate_full_name(gender)
        last_name = fake.last_name()
        second_last_name = fake.last_name()

        full_name = f"{full_name} {last_name} {second_last_name}"
        id_hash = hashlib.md5(full_name.encode()).hexdigest()
        phone_number = "5255" + str(random.randint(0, 999999)).zfill(6)
        public_hash = hashlib.sha256(full_name.encode()).hexdigest()
        private_hash = hashlib.sha512(full_name.encode()).hexdigest()
        value = random.randint(1, 5)
        timestamp = random.randint(0, int(time.mktime(current_date.timetuple())))
        timestamp = datetime.fromtimestamp(timestamp).strftime('%Y%m%d')
        clave_elector = str(random.randint(0, 9999999999999)).zfill(13)

        data = {
            "id": id_counter,
            "name": full_name,
            "id_hash": id_hash,
            "phone": phone_number,
            "public_hash": public_hash,
            "private_hash": private_hash,
            "value": value,
            "timestamp": timestamp,
            "clave_elector": clave_elector
        }
        dataset.append(data)
        id_counter += 1
    return dataset

dataset = generate_dataset()

for data in dataset[:100]:
    print(data)
```
