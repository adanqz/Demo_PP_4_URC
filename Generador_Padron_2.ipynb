```python
import random
import hashlib
import datetime


nombres_femeninos = ['Karla', 'Sofia', 'America', 'Daniela', 'Ariana', 'Jocelyn', 'Andrea', 'Monica', 'Monserrath', 'Karina']
nombres_masculinos = ['Roberto', 'Miguel', 'Ruben', 'Daniel', 'Adan', 'Jose', 'Rafael', 'Maximiliano', 'Raul', 'Isaac', 'Manuel', 'Daniel', 'Carlos', 'Jared', 'Enrique', 'Jesus', 'Ricardo', 'Sergio', 'Alejandro', 'Victor', 'Francisco' , 'Angel', 'David', 'Abraham']
apellidos = ['Garcia', 'Hernandez', 'Lopez', 'Martinez', 'Gonzalez', 'Rodriguez', 'Carmona', 'Quintero', 'Sánchez', 'Miranda', 'Garcia', 'Campero', 'Espinosa', 'Casas', 'Morales', 'Napoles', 'Romero', 'Gordillo', 'Guerra', 'Rojas', 'Valdes', 'Perez', 'Contreras', 'de la Cruz', 'Flores', 'Rincon', 'Ramirez', 'Jaimes', 'Castro', 'Evangelista', 'Buendia', 'Izquierdo', 'Angeles', 'Pineda']

dataset = []

for i in range(9000):

    id = i

    genero = random.choice(['F', 'M'])
    if genero == 'F':
        if random.choice([True, False]):
            nombre = random.choice(nombres_femeninos)
        else:
            nombre = random.choice(nombres_femeninos) + ' ' + random.choice(nombres_femeninos)
    else:
        if random.choice([True, False]):
            nombre = random.choice(nombres_masculinos)
        else:
            nombre = random.choice(nombres_masculinos) + ' ' + random.choice(nombres_masculinos)

    apellido1 = random.choice(apellidos)
    apellido2 = random.choice(apellidos)

    telefono = '5255' + ''.join(random.choice('0123456789') for _ in range(6))

    hash_publico = hashlib.sha256(str(random.random()).encode()).hexdigest()
    hash_privado = hashlib.sha256(str(random.random()).encode()).hexdigest()

    hash_dependiente = hashlib.sha256(hash_privado.encode()).hexdigest()

    campo_valores = random.randint(1, 5)

    fecha_actual = datetime.datetime.now().date()
    fecha_aleatoria = datetime.date.fromordinal(random.randint(1, fecha_actual.toordinal()))
    timestamp = fecha_aleatoria.strftime('%Y%m%d')

    clave_elector = ''.join(random.choice('0123456789') for _ in range(13))

    registro = {
        'ID': id,
        'Nombre': nombre,
        'Apellido1': apellido1,
        'Apellido2': apellido2,
        'Teléfono': telefono,
        'Hash Público': hash_publico,
        'Hash Privado': hash_privado,
        'Hash Dependiente': hash_dependiente,
        'Campo de Valores': campo_valores,
        'Timestamp': timestamp,
        'Clave Electoral': clave_elector
    }
    dataset.append(registro)

for registro in dataset:
    print(registro)
```
